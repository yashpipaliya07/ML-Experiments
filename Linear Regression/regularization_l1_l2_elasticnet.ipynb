```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("train.csv")

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

X = (X - X.mean(axis=0)) / X.std(axis=0)

n, m = X.shape

X = np.c_[np.ones(n), X]
```

# L1 Regularization lasso
J = MSE + λ ∑∣θ∣

```python
theta_l1 = np.zeros(m + 1)
lr = 0.01
lambda_l1 = 0.1
epochs = 1000

for _ in range(epochs):
    y_pred = X @ theta_l1
    error = y_pred - y

    gradient = (1/n) * (X.T @ error)

    l1 = lambda_l1 * np.sign(theta_l1)  # only takes the sign of the l1 not whole l1 cause it is derrivative of mse
    l1[0] = 0  # no regularization for bias

    theta_l1 -= lr * (gradient + l1)

print("L1 (Lasso) Coefficients:\n", theta_l1)
```

# L2 Regularization Ridge

J = MSE + λ ∑θ^2


```python
theta_l2 = np.zeros(m + 1)
lr = 0.01
lambda_l2 = 0.1
epochs = 1000

for _ in range(epochs):
    y_pred = X @ theta_l2
    error = y_pred - y

    gradient = (1/n) * (X.T @ error)

    l2 = lambda_l2 * theta_l2
    l2[0] = 0  # no regularization for bias

    theta_l2 -= lr * (gradient + l2)

print("L2 (Ridge) Coefficients:\n", theta_l2)
```

# Regularization Elastic Net

J = MSE + λ1 ​∣θ∣ + λ2 ​θ^2

```python
theta_enet = np.zeros(m + 1)
lr = 0.01
lambda_l1 = 0.1
lambda_l2 = 0.1
epochs = 1000

for _ in range(epochs):
    y_pred = X @ theta_enet
    error = y_pred - y

    gradient = (1/n) * (X.T @ error)

    l1 = lambda_l1 * np.sign(theta_enet)
    l1[0] = 0

    l2 = lambda_l2 * theta_enet
    l2[0] = 0

    theta_enet -= lr * (gradient + l1 + l2)

print("Elastic Net Coefficients:\n", theta_enet)
```

```python
import numpy as np
import pandas as pd
```

# Preparing the Data

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

train_data = train_data.drop("ID", axis=1)
test_data = test_data.drop("ID", axis=1)

x_train = train_data.drop("medv", axis=1).values
y_train = train_data["medv"].values

x_test = test_data.values
#our test data set does not contains medv (result) means we dont have y test value

# feature scaling

mean = x_train.mean(axis=0)
std = x_train.std(axis=0)

std[std == 0] = 1

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std

x_train = np.c_[np.ones(x_train.shape[0]),x_train]
x_test = np.c_[np.ones(x_test.shape[0]),x_test]
```

MSE Function

```python
def mse(x,y,theta):
  return np.mean((x @ theta - y)**2)
```

# Batch GD

```python
theta_batch = np.zeros(x_train.shape[1])
lr = 0.01
epochs =1000
n= x_train.shape[0]

for _ in range(epochs):
  y_pred = x_train @ theta_batch
  gradient = (1/n) * (x_train.T @ (y_pred-y_train))
  theta_batch = theta_batch - lr * gradient
```

#Stochastic GD

```python
theta_sgd = np.zeros(x_train.shape[1])
lr = 0.01
epochs = 50
n = x_train.shape[0]

for _ in range(epochs):
  for(i) in range(n):
    xi = x_train[i]
    yi = y_train[i]

    y_pred = xi @ theta_sgd
    gradient = (y_pred-yi) * xi
    theta_sgd = theta_sgd - lr*gradient
```

#Mini_Batch GD

```python
theta_mb = np.zeros(x_train.shape[1])
lr = 0.01
epochs = 200
batch_size = 32
n = x_train.shape[0]

for _ in range(epochs):
  indices = np.random.permutation(n)
  x_shuf = x_train[indices]
  y_shuf = y_train[indices]

  for i in range(0,n,batch_size):
    x_batch = x_shuf[i:i+batch_size]
    y_batch = y_shuf[i:i+batch_size]

    y_pred = x_batch @ theta_mb
    gradient = (1/len(x_batch)) * (x_batch.T @ (y_pred - y_batch))
    theta_mb = theta_mb - lr * gradient

```

#Normal Equation


```python
theta_normal = np.linalg.inv(x_train.T @ x_train) @ x_train.T @ y_train
```

#Evaluation

```python
pred_batch = x_test @ theta_batch
pred_sgd = x_test @ theta_sgd
pred_mb = x_test @ theta_mb
pred_normal = x_test @ theta_normal


print("===== TRAIN MSE =====")
print("Batch GD:", mse(x_train, y_train, theta_batch))
print("SGD:", mse(x_train, y_train, theta_sgd))
print("Mini-Batch GD:", mse(x_train, y_train, theta_mb))
print("Normal Equation:", mse(x_train, y_train, theta_normal))

print("\n===== SAMPLE TEST PREDICTIONS =====")
print("Batch GD:", pred_batch[:5])
print("SGD:", pred_sgd[:5])
print("Mini-Batch GD:", pred_mb[:5])
print("Normal Equation:", pred_normal[:5])
```
